```python
# import the module and open the source data and view the top data
import pandas as pd
hd=pd.read_csv('https://www.dropbox.com/s/ltksjtb54wrov5q/HeartDisease.csv?dl=1')
hd.head()
```

```python
# see the information about the data set
hd.info()
```

```python
# statistical report
hd.describe()
```

```python
# show the columns
hd.columns
```

```python
# defining X & Y
y=hd['target']
X=hd[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
```

```python
# split dataset into train & test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2069)
```

```python
# select the ML model <KNN Classifier>
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
# fitting
knc.fit(X_train,y_train)
```

```python
# predict the model
y_pred=knc.predict(X_test)
```

```python
# checking accuracy
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
#Matrix form
confusion_matrix(y_test,y_pred)
```
